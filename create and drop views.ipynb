```python
import pandas as pd
import sqlite3 as sql
```

```python
database = "chinook.db"
conn = sql.connect(database)
```

```python
def query(q):
    return pd.read_sql_query(q,conn)
```

### calling saved views

```python
q = """ select * from sqlite_master where type = "view" """
query(q)
```

```python
q = """select * from Regular_Customer"""
query(q)
```

### Creating view

```python
q = """ create view Emloyee_information as
    select e.first_name || " " || e.last_name Name,e.title Designation,e.country Country, 
    e2.first_name || " " || e.last_name "Reports To"
    from employee e 
    left join employee e2
    on 
    e.reports_to = e2.employee_id"""
conn.execute(q)
```

### check, is view create or not

```python
q = """select * from sqlite_master where type = "view"  """
query(q)
```

```python
q = """ select * from Emloyee_information"""
query(q)
```
